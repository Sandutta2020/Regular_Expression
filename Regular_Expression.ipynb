```python
import re
```

```python
list_email =["santan0u@yahoo.com","Santanu@gmail.com","invalid@gmail","invalid@com","santanugmail","santanu@yahoo.in"]
```

```python
## check email address have "@" or not
for items in list_email:
    matchObj = re.search( '@', items)
    if matchObj:
        print(items)
```

```python
for items in list_email:
    matchObj = re.search(r'([\w\.-]+)@([\w\.-]+)', items)
    if matchObj:
        print("Email address:", matchObj.group()) # The whole matched text
        print("Username:", matchObj.group(1)) # The username (group 1)
        print("Host:", matchObj.group(2)) # The host (group 2)
```

```python
for items in list_email:
    matchObj = re.search(r'([\w\.-]+)@([\w]+)\.([\w]+)', items)
    if matchObj:
        print("Email address:", matchObj.group()) # The whole matched text
        print("Username:", matchObj.group(1)) # The username (group 1)
        print("Host:", matchObj.group(2)) # The host (group 2)
        print("Host:", matchObj.group(3)) # The host (group 2)
```

```python
#only com and org is allowed

for items in list_email:
    matchObj = re.search(r'([\w\.-]+)@([\w]+)\.(com|org)', items)
    if matchObj:
        print("Email address:", matchObj.group()) # The whole matched text
        print("Username:", matchObj.group(1)) # The username (group 1)
        print("Host:", matchObj.group(2)) # The host (group 2)
        print("Host:", matchObj.group(3)) # The host (group 2)
```

```python
movies =["Shawshank Redemption, The (1994)","Schindler's List (1993)","Sixth Sense, The (1999)","Blade Runner (1982)","One Flew Over the Cuckoo's Nest (1975)","Saving Private Ryan (1998)"]
```

```python
for items in movies:
    matchObj = re.search(r'\((.*)\)', items)
    matchObj2 = re.search(r'(.*)\((.*)\)', items)
    if matchObj:
        print("Movie Release:", matchObj.group(1)) # 
    if matchObj2:
        print("Movie Name:", matchObj2.group(1),matchObj2.group(2)) # 
    
```

```python
statement = "Please contact us at: support@datacamp.com, xyz@datacamp.com"

#'addresses' is a list that stores all the possible match
addresses = re.findall(r'[\w\.-]+@[\w\.-]+', statement)
for address in addresses:
    print(address)
```

```python
statement = "12345 san234 santanu SANTANU   ssss @#%%"
```

```python
addresses_numeric = re.findall(r'[\d]+', statement)
for address in addresses_numeric:
    print(address)
print('-------------------------------')
addresses_alpha = re.findall(r'[\D]+', statement)
for address in addresses_alpha:
    print(address)
print('-------------------------------')
addresses_alpha_numeric = re.findall(r'[\w]+', statement)
for address in addresses_alpha_numeric:
    print(address)
```

```python
# find string starts with The  and ends with list
movies =["The Shawshank Redemption","The Schindler list","My list", "The Plainlist","my listtt","Thelist"]

for items in movies:
    x = re.search("^The.*list$", items)
    if x:
        print(x.group())
```

```python
# find string starts with only vowel
words =["almond","Easy","cloud", "violet","Ink","The"]
vowel =[]
consonant =[]
for items in words:
    words_start_with_vowel = re.search("^[aeiouAEIOU].*", items)
    words_start_with_constant = re.search("^[^aeiouAEIOU].*", items)
    if words_start_with_vowel:
        vowel.append(words_start_with_vowel.group())
    if words_start_with_constant:
        consonant.append(words_start_with_constant.group())
print(vowel,consonant)
```
